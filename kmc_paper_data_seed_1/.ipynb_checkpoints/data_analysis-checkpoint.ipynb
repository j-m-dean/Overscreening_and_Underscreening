```python
from copy import deepcopy
def calculate_exclusion(distance, rdf):
    d = 0
    for i in range(len(distance)):
        if rdf[i] == 0:
            if distance[i] > d:
                d = distance[i]
                
    ordered_distance = list(deepcopy(distance))
    ordered_distance.sort()
    if d > 0.:
        idx = ordered_distance.index(d)
        return(ordered_distance[idx])
    else:
        try:
            idx = ordered_distance.index(d)
            return(ordered_distance[idx])
        except ValueError:
            return(2.5e-10)
     
```

```python
def yukawa_osc(x,alpha, A, d, b, c):
    return(d-A*np.exp(-alpha*x)/x * np.cos(b*x + c))
```

```python
def func_osc(x, alpha, A,f, b, c):
    return(f - A*np.exp(-alpha * x)*np.cos(b*x+c))

def func_exp(x, alpha, A,f):
    return(f - A*np.exp(-alpha * x))
```

```python
def debye(occ, perm):
    num = perm * 8.854e-12 * 1.381e-23 * 300
    conc = occ / (75 * 2.5e-10)**3
    denom = conc * 1.602e-19**2
    return(np.sqrt(num / denom))

perms = np.array([1,4,7,10,13,16,19,22,25,28,50,75,100])

debye_lengths_105 = debye(105, perms)
debye_lengths_210 = debye(210, perms)
debye_lengths_421 = debye(421, perms)

x_105 = np.log(ex_105/ debye_lengths_105)
x_210 = np.log(ex_210/ debye_lengths_210)
x_421 = np.log(ex_421/ debye_lengths_421)
```

```python
def debye(occ, perm):
    num = perm * 8.854e-12 * 1.381e-23 * 300
    conc = occ / (75 * 2.5e-10)**3
    denom = conc * 1.602e-19**2
    return(np.sqrt(num / denom))

```

```python
N = 31
mid = int(0.5*(N-1))

r_values = {}

for i in range(-mid, mid+1):
    for j in range(-mid, mid+1):
        for k in range(-mid, mid+1):
            r = np.linalg.norm(np.array([i,j,k]))
            if r in r_values.keys():
                if j in r_values[r].keys():
                    r_values[r][j] += 1
                else:
                    r_values[r][j] = 1
            else: 
                r_values[r] ={}
                r_values[r][j] = 1
                
r_values             
```

```python
c_bulk =  0.001
c = np.ones(N) * c_bulk
```

```python
def beta_w(bjerrum, a, r, kappa):
    num = bjerrum * np.exp(- kappa * a * r)
    denom = a * r
    return num / denom 

c[15] = 0.005

def correlation_term(bjerrum, a, r, kappa):
    
    return np.exp( -  beta_w(bjerrum, a, r, kappa)) - 1

def correlations(c, rs, bjerrum, a):
    corre = []
    for i in range(len(c)):
        c = np.roll(c,1)
        v = 0
        mid = int(0.5*(len(c) - 1))
        pre_kappa = np.sqrt(4 * np.pi * bjerrum / a**3)

        for r in rs.keys():
            if r != 0.0:
                for z in rs[r].keys():
                    kappa = pre_kappa * np.sqrt(c[mid + z]*(1 - c[mid + z]))
                    v += rs[r][z] * correlation_term(bjerrum, a, r, kappa)
                v = - (2 * np.pi * bjerrum / (a * r)) * v
        corre.append(v)
    print(corre)
```
